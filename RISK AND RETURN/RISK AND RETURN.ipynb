```python
import yfinance as yf
import riskfolio as rp
import warnings
warnings.filterwarnings("ignore")
```

```python
assets = ["AKBNK.IS", "AKSEN.IS", "ARCLK.IS", "BIMAS.IS", "TUPRS.IS", "EREGL.IS", "PETKM.IS", "THYAO.IS",
"GARAN.IS", "YKBNK.IS", "ISCTR.IS", "PETKM.IS", "PGSUS.IS", "ASELS.IS", "SISE.IS", "KCHOL.IS"]
```

```python
assets.sort()
```

```python
data = yf.download(assets, start="2010-01-01", end="2022-01-12")
returns = data['Adj Close'].pct_change().dropna()
```

```python
port = rp.Portfolio(returns=returns)
```

```python
port.assets_stats(method_mu='hist', method_cov='hist', d=0.94)
```

```python
w_rp = port.rp_optimization(
model="Classic",
rm="MV",
hist=True,
rf=0,
b=None
)
```

```python
ax = rp.plot_pie(w=w_rp)
```

```python
ax = rp.plot_risk_con(
w_rp,
cov=port.cov,
returns=port.returns,
rm="MV",
rf=0,
)
```
